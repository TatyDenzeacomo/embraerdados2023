```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {"Nome": ['Alice', 'João', 'Joana', 'Ana', 'Gustavo'],
        "Idade": [25, 30, 45, 23, 50],
        "Salario": [5000, 2000, 7000, 11000, 8000]}
```

```python
df = pd.DataFrame(data)
```

```python
# Plotar um gráfico de barras para salários
plt.bar(df['Nome'], df['Salario'], color='blue')
plt.title('Grafico de Barras = Salario')

plt.xlabel('Nome das pessoas')
plt.ylabel('Salário (R$)')
plt.show()
```

```python
# gráfico de pizza

plt.pie(df['Salario'], labels=df['Nome'], autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink'])

plt.title('Distribuição de Salários')
plt.show()
```

```python
# grafico de dispersão
plt.scatter(df['Idade'], df['Salario'], color='purple')
plt.title('Grafico de Dispersão = Idade vs. Salário')
plt.xlabel('Idade')
plt.ylabel('Salario (R$)')
plt.show()
```

```python
# Gráfico de linhas para Idades

plt.plot(df['Nome'], df['Idade'], marker='o', linestyle='-', color='green')
plt.title('Gráfico de Linhas = Idade')
plt.xlabel('Nome')
plt.ylabel('Idade')
plt.show()
```

```python
# Gráfico de linhas para Salário

plt.plot(df['Nome'], df['Idade'], marker='X', linestyle='-', color='green')
plt.title('Gráfico de Linhas = Idade')
plt.xlabel('Nome')
plt.ylabel('Salario')
plt.show()
```
